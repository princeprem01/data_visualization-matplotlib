```python
fig, ax=plt.subplots(nrows=1, ncols=2, figsize=(12,6))
a1=df.plot(x='age', y='chol', kind='scatter', c='target', cmap='BrBG', ax=ax[0]);
df.plot(y='age', kind='hist', bins=40, ax=ax[1]);
a1.legend(*a1.collections[0].legend_elements(), title='Legend') #The ax.legend() call does not automatically generate a legend for a scatter plot based on the c (color) argument because matplotlib doesn't inherently know how to interpret the c values as categories for the legend. The legend needs to be explicitly told what each color represents.
```

```python
#2 OO method:Lets do some more customizations

fig, ax= plt.subplots(figsize=(8,5))
sc=ax.scatter((df['age']),df['chol'],c=df['target'], cmap='BrBG');
ax.set(xlabel='age', ylabel='chol');
ax.legend(*sc.legend_elements(), title="Legends"); #The ax.legend() call does not automatically generate a legend for a scatter plot based on the c (color) argument because matplotlib doesn't inherently know how to interpret the c values as categories for the legend. The legend needs to be explicitly told what each color represents.
```

```python
fig, ax= plt.subplots(figsize=(8,5))
axes=ax.scatter((df['age']),df['chol'],c=df['target'], cmap='BrBG');
axes.legend_elements()
```

```python
fig, ax= plt.subplots(figsize=(8,5))
sc=ax.scatter((df['age']),df['chol'],c=df['target'], cmap='BrBG');
ax.set(xlabel='age', ylabel='chol')
ax.axhline(df.chol.mean(), linestyle='--', c='black' )
ax.legend(*sc.legend_elements(), title="Legends");
#added a horizontal line
```

```python
fig, ax= plt.subplots(ncols=1,nrows=2,figsize=(9,9), sharex=True)

a1=ax[0].scatter(df['age'],df['chol'], c=df['target'], cmap='BrBG') #df.plot(df['age'], df['chol'] KeyError: "None of [Index([63, 37, 41, 56, 57, 57, 56, 44, 52, 57,\n       ...\n       67, 44, 63, 63, 59, 57, 45, 68, 57, 57],\n      dtype='int64', length=303)] are in the [columns]". df is already in the begaining so just provide column names and not df. or again df[''] 
ax[0].set(title='Age vs Chol')
ax[0].axhline(df['chol'].mean(), linestyle='--')
ax[0].legend(*a1.legend_elements(),title='Legend1') #using pandas' DataFrame.plot with kind='scatter', adding a legend for the c (color) argument requires a bit of manual work because pandas does not automatically generate legends

a2=ax[1].scatter(df['age'],df['thalach'], c=df['target'], cmap='Accent_r');
ax[1].set(title='Age vs Thalach')
ax[1].axhline(df.thalach.median(), linestyle='--');
ax[1].legend(*a2.legend_elements(), title='Legend2')

fig.suptitle('Scatter Graphs', fontsize=14, fontweight='bold'); #complete figure's super title
```
