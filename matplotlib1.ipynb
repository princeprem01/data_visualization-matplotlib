```python
#TWO WAYS OF PLOT 
x=[20,30,40,50]
y=[50,40,30,20]

#1st way
fig=plt.figure() #creates a figure
ax=fig.add_subplot() #add axes as subplot
ax.plot(x,y,marker='o');
```

```python
#2nd way(BETTER)
fig, ax= plt.subplots() #simultaneously draws figure & axes
ax.plot(x,y,marker='o');
```

```python
fig, ax= plt.subplots();
type(fig), type(ax) #the pyplot API is generally less-flexible than the Object-oriented API(BETTER).
```

```python
X=np.arange(0,6, 1)

fig, ax= plt.subplots()
ax.plot(X, X**2, label='X**2') #No legend in this as such
ax.plot(X**2, X, label='Y**2') #you can use label to Specifies the text for each line in the legend.
ax.set(xlim=(0,10), ylim=(0,10))
ax.legend();
```

## Example of workflow for matplotlib

```python
#import matplotlib.pyplot ..

#Prepare the data
x,y=[20,30,40,50],[50,40,30,20]

#setup fig & axes
fig, ax= plt.subplots(figsize=(5,3)) #figure size= (width, height)

#put & plot the data
ax.plot(x,y,marker='o');

#customize the plot
ax.set(title='Example plot', xlabel='x-axis', ylabel='y-axis', xlim=(10,100), ylim=(10,100));

# plt->figure->axes/subplot->plot data-> .set()
```

```python
#Plot with dictionary data
data=dict({"Bangalore": 40,
           "Mumbai": 50,
           "Ahemdabad":60,
           "Gujrat": 30,
           "Pune": 20    
})
data
```

```python
data.keys(), type(data.values())
```

```python
fig, ax= plt.subplots()
#ax.bar( x=data.keys(),height=data.values())
ax.bar(data.keys(),height=data.values())
ax.set(xlabel='Cities', ylabel='Costs', title='BAR GRAPH');
```

```python
np.array(data.values())
list(data.values()) #barh; TypeError: unhashable type: 'dict_keys' so turn them into float or array-like
```

```python
fig,ax= plt.subplots()
ax.barh(list(data.keys()), width=list(data.values())); #y,width= x,y
ax.set(ylabel='Cities', xlabel='Costs', title='BAR GRAPH');
```

```python
fig, (ax0, ax1)= plt.subplots(nrows=1, ncols=2, figsize=(15,4)) #nrows= no of rows, ncols
ax0.bar(data.keys(),data.values())
ax1.barh(list(data.keys()),list(data.values()));
```

```python
fig, (ax0, ax1, ax3)= plt.subplots(nrows=1, ncols=3, figsize=(6,4))
```

```python
fig, ax= plt.subplots(nrows=3, ncols=1, figsize=(4,6))
ax[0].plot(x, np.log(x)) #ax[0,0]; the error occurs because the ax object created by plt.subplots is a 1-dimensional array when nrows=3 and ncols=1. In this case, youre indexing it as if it were a 2D array (ax[0,1]), which is invalid.
ax[1].scatter(x, np.exp(x))
ax[2].barh(list(data.keys()), list(data.values()));
```

```python
fig, ((ax0, ax1), (ax2, ax3))= plt.subplots(nrows=2, ncols=2, figsize=(15,8))
ax0.plot(x, x**2)
ax1.plot(x, np.log(x))
ax2.scatter(x,np.exp(x))
ax3.bar(data.values(), data.keys())
ax0.set(title="simple plot 1")
ax1.set(title="simple plot 2")
ax2.set(title="scatter plot")
ax3.set(title="bar plot");
```

```python
fig, ax= plt.subplots(nrows=2, ncols=2, figsize=(15,8))
ax[0,0].plot(x, x**2)
ax[0,1].plot(x, np.log(x))
ax[1,0].scatter(x,np.exp(x))
ax[1,1].bar(data.values(), data.keys());
```
