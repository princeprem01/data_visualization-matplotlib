```python
df['Date']= pd.date_range('1/1/2003', periods=len(df))
df
```

```python
df.info()
```

```python
(df['Price'].str.replace('$', '').str.replace(',', '')).str[:-3]
```

```python
#Or use Regex ie Regular expression| https://regexone.com/lesson
(df['Price'].str.replace('\D','', regex=True) ).str[:-2] #\d for all numbers
(df['Price'].str.replace('[$,\.]','', regex=True) ).str[:-2]
```

```python
df.Price=(df['Price'].str.replace('[$,\.]','', regex=True) ).str[:-2]
df
```

```python
df.Price=df['Price'].astype(int)
```

```python
df.info()
```

```python
df['Total_Profit']= df.Price.cumsum() #df.Tota_Profit=.. UserWarning: Pandas doesn't allow columns to be created via a new attribute name. 
df
```

```python
df.drop(columns=['Tota_Profit'],axis=1, inplace=True)
```

```python
df
```

```python
df.plot(x='Date', y='Total_Profit')
```

```python
df.plot(x='Date', y='Price', kind='scatter', figsize=(15,5));
```
